# [Step 2] 시간 축(Timeline) 기준 데이터 분할 (Train / Valid / Test)

```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path(".")
DATA_RAW = BASE_DIR / "data" / "raw"

# 시계열 순서가 파괴되어 미래 정보가 과거 학습에 개입하는 Data Leakage 원천 차단
raw_file_path = DATA_RAW / "raw_ohlcv_100.csv"
total_df = pd.read_csv(raw_file_path)
total_df['date'] = pd.to_datetime(total_df['date'])

unique_dates = sorted(total_df['date'].unique())
total_days = len(unique_dates)

train_end_idx = int(total_days * 0.8)
valid_end_idx = int(total_days * 0.9)

train_dates = unique_dates[:train_end_idx]
valid_dates = unique_dates[train_end_idx:valid_end_idx]
test_dates = unique_dates[valid_end_idx:]

train_df = total_df[total_df['date'].isin(train_dates)].copy()
valid_df = total_df[total_df['date'].isin(valid_dates)].copy()
test_df = total_df[total_df['date'].isin(test_dates)].copy()

train_df.to_csv(DATA_RAW / "train.csv", index=False, encoding="utf-8-sig")
valid_df.to_csv(DATA_RAW / "valid.csv", index=False, encoding="utf-8-sig")
test_df.to_csv(DATA_RAW / "test.csv", index=False, encoding="utf-8-sig")

print(f"[+] 데이터셋 분할 완료 (Train: {len(train_df)}행, Valid: {len(valid_df)}행, Test: {len(test_df)}행)")
```
